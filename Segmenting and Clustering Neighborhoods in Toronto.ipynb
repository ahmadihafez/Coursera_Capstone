# Question one (Target a table of Canada postal codes starting with M)

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
df=df[df['Borough'] != "Not assigned" ]
df.reset_index(drop=True)
```

# Question 2 (Target Table + Lat. and Long.)

```python
df_pc=pd.read_csv("https://cocl.us/Geospatial_data")
```

```python
df_pc
```

```python
df_PC_B_N_co = pd.merge(df,  
                      df_pc,  
                      on ='Postal Code',  
                      how ='inner') 
```

```python
df_PC_B_N_co
```

# Maps to visualize your neighborhoods

```python
#Visualizing all the Neighbourhoods of the above data frame using Folium
import folium
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df_PC_B_N_co['Latitude'],df_PC_B_N_co['Longitude'],df_PC_B_N_co['Borough'],df_PC_B_N_co['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=3,
    popup=label,
    color='red',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.5,
    parse_html=False).add_to(map_toronto)
map_toronto
```
